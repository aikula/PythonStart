# Обработка исключений
Немного о синтаксисе. Вложенные блоки в Python выднеляются четырьями пробелами. И строка, которая открывает блок заканчивается двоеточием. Например:

```python
try:
    #вложенный блок
    D3=float(S4)
    print('Вложенный блок')
except:
    #второй вложенный блок
    print('Вы ввели не числовое значение')
print('Код без вложения')
```

Если между в коде между try и except возникает ошибка (более правильно говорить исключение), то код не прекратит выполнение. Но будет выполнен код блока except. Те у нас есть возможность обработать исключение. Например, сообщить пользователю, что он ввел не верное значение.

```python
S4=input('Введите число: ')
try:
    D3=float(S4)
except:
    D3=None
    print('Вы ввели не число.')
print(D3)
```

# Циклы и ветвления
Давайте представим, что нам надо выполнять какую-то операцию несколько раз. Здесь нам пригодятся циклы, те конструкции, которые могут выполняться указанное количество раз. Проще говоря, повторение. Самый простой вид цикла - while.

```python
i=1 #переменная будет у нас выполнять подсчет выпленных циклов
while True: #исполнять пока истина, фактически это вечный цикл
    print('Выполняю повторение: ', i)
    i+=1 #подсчитываем количество выболненных циклов. Увеличиваем i на единицу. Эквивалентно i=i+1
```

Чтобы выйти из такого цикла можно использовать оператор break, он прервет исполнения цикла. А понять, когда надо прервать нам помогут операторы ветвления. В частности конструкция if ... then ... else...

```python
D5=input('Введите слово "корова": ') 
if D5=='корова': #обратите внимание, что логический оператор сравнения это '==', оператор присваивания '='
                 #Python в именах переменных и строках различает большие и малые буквы и "Корова" не будет равно "корова"
    print('У вас все получилось! Поздравляем.')
else:
    print('Вы ошиблись в написании')
```

```python
#Давайте вспомним булевые переменные, которые принимают значение истина (True) или ложь (False)
K1=1
K2=3
print('Равенства K1==1', K1==1)
print('Меньше K1<K2', K1<K2) 
print('Больше K1>K2', K1>K2)
print('Больше или равно K1>=K2', K1>=K2)
print('Не равно K1!=K2', K1!=K2 )
print('Цепочка условий 0<K1<10', 0<K1<10)
```

Вернемся к "вечному циклу".

```python
i=1 #переменная будет у нас выполнять подсчет выпленных циклов
while True: #исполнять пока истина, фактически это вечный цикл
    print('Выполняю повторение: ', i)
    if i==10:
        break
    else:
        i+=1 #подсчитываем количество выболненных циклов. Увеличиваем i на единицу. Эквивалентно i=i+1
```

```python
for i in range(1,11): #генерирует последовательность от 1 до 10
    print(i)
```

# Практическая задача

Давайте решим простую задачу из имитационного моделирования. Например, нам надо определить при каком остатке товаров оптимально сделать заказ и какого размера должен быть заказ. При этом:

- количество приобретаемого товара в день описывается нормальным распределением с ожидаемым значением 10 единиц и стандартным отклонением 2 единицы (не пугайтесь математики :)). Это означает, что в 95% случаев спрос будет лежать в диапазоне от 6 до 14 единиц, но чаще ближе к 10 единицам.
- срок поставки заказа составляет 2 дня
- при продаже товара мы зарабатываем 100 рублей
- хранение товара нам обходится в 30 рублей в день
- стартовый запас 40 единиц товара.
Наша задача расчитать, при каком остатке товара на складе оптимально делать заказ и размер заказа. Период на котором будем проводить моделирование - 30 дней.

Алгоритм будет выглядеть следующим образом:
1. Создадим двухмерный пустой массив (заполним нулями). В первом столбце будем хранить остаток товара, во втором - поставленный товар.
3. Получим от пользователя значение, при котором будем делать заказ и размер заказа
2. Запустим цикл в рамках которого будем:
- моделировать спрос
- если достаточно товара - отпускать его клиентам и подсчитывать прибыль, в противном случае будем отпускать только имеющийся товар
- проверять надо ли сделать заказ

```python
#импортируем библиотеку, которая позволит моделировать гаусовское распределение
from random import gauss

#стартовые значения
n=30 #количество периодов для моделирования
profitOne=100  #прибыль на единице товара
cost=30 #затраты на зранение единицы товара 
dayIn=2 #через сколько дней будет поставлен товар
profitSumm=0 #инициируем переменную, которая будет аккумулировать прибыль

#запрашиваем данные от пользователя, используем написанную нами ранее функцию
print('Введите объем заказа: ')
zak=inputint()
print('Введите минимальный запас: ')
minZap=inputint()

#инициируем двух мерный массив, где первый столбец будет хранить актуальный запас, 
#а второй, поставку для соответсвующего периода
matrix=[[0 for a in range(0,n+2)],[0 for a in range(0,n+2)]]
matrix[0][0]=40 #стартовый запас товара

#запускаем цикл 0 до n
for i in range(n):
    profitSumm-=matrix[0][i]*cost #рассчитываем стоимость хранения
    matrix[0][i]+=matrix[1][i] #добавляем объем поставленного товара в этот период
    spros=int(gauss(10,2)) #моделируем спрос, используем функцию нормального распределения
    
    #проверяем, сколько товара можем отгрузить не вохникает ли дефицит
    if matrix[0][i]>=spros:
        #если дефицита нет, то отгружаем товар в рамках спроса
        matrix[0][i+1]=matrix[0][i]-spros 
        profitSumm+=spros*profitOne #подсчитываем прибыль
        noStock=0 #указываем, что дефицита не возникло
    else: #товара на складе меньше, чем был спрос
        profitSumm+=matrix[0][i]*profitOne #считаем прибыль исходя из продажи товара, который был
        matrix[0][i+1]=0 #закас на складе в конце периода 0
        noStock=spros-matrix[0][i] #рассчитываем объем неудовлетворенного спроса
        
    #в рамках этого условия проверим, надо ли сделать заказ
    #и сразу есть второе условие: проверяем, не выйдет ли срок заказа за 31 день. Так как у нас массив сгенерирован
    #на 31 период, то например, если в 30 периоде надо будет делать заказ, программа обратиться к ячейке 30+2 - а ее не 
    #существует. Воникнет исключение
    if matrix[0][i]<minZap and (i+dayIn)<n:
        matrix[1][i+dayIn]=zak
    
    #выведим результы для периода
    print('период: ', i, ' спрос: ', spros, 'Запас на начало периода: ', matrix[0][i]+matrix[1][i], 'Остаток: ', matrix[0][i+1], 'Дефицит: ',  noStock)
#вне цикла выведем итоговую прибыль за весь период
print('Прибыль составила: ', profitSumm)        
```
